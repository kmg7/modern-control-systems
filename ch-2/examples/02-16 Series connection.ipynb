```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
import control as ct

# Set matplotlib style for better plots
plt.style.use('default')
plt.rcParams['figure.figsize'] = (10, 8)
plt.rcParams['grid.alpha'] = 0.3
```

```python
sysg = ct.tf([1], [500, 0, 0])
print(sysg)
```

```python
sysgc = ct.tf([1, 1], [1, 2])
print(sysgc)
```

```python
sys3 = ct.series(sysg, sysgc)
print(sys3)
```

### The parallel one

```python
print(ct.parallel(sysg, sysgc))
```

### Closing the loop with with a unity feedback signal

```python
sign = -1  # Negative feedback (default)
sys = ct.feedback(sys3, [1], sign) # 1 means unity feedback
print(sys)
```
