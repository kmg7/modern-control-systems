```python
import numpy as np
import control as ct

p = [1,7,10]
q = [1,2]

polP = np.poly1d(p, variable='s')
polQ = np.poly1d(q, variable='s')

print(polP, polQ, sep='\n')
```

```python
# a
pq = polP * polQ

print(pq)
```

```python
# b
sysG = ct.tf(p, q)

print(sysG)
print("poles", sysG.poles(), "zeros", sysG.zeros(), sep='\n')
# print(sysQ)
```
