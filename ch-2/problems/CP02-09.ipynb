```python
import control as ct
import numpy as np

sys = ct.series(ct.tf([1, 1], [1, 2]), ct.tf([1], [1, 1]))
```

```python
# a
print(sys)
```

```python
# b
pzmap = ct.pole_zero_map(sys)
print("Poles:", pzmap.poles)
print("Zeros:", pzmap.zeros)
pzmap.plot()
```

```python
# c
sys_min = ct.minreal(sys)
ct.pole_zero_map(sys_min).plot()
```

A pole is where the system at barely infinite, and a zero is at zero. If we think of our system analysis in s plane like a 3d geographical map the poles stand up for mountains that almost impossible to reach. Then zeros will be at sea level or core of our planet.
