```python
import control as ct
import numpy as np
import display_utils as dp
import matplotlib.pyplot as plt
```

```python
tf_c = ct.tf([3], [1, 3])
dp.tf(tf_c)
tf_p = ct.tf([1], [1, 2, 5])
dp.tf(tf_p)
```

```python
ss_c = ct.ss(tf_c)
dp.ss(ss_c)
ss_p = ct.ss(tf_p)
dp.ss(ss_p)
```

```python
sys_cp = ct.series(ss_c, ss_p)
sys = ct.feedback(sys_cp, 1)
dp.ss(sys)

T, y_out = sys.impulse_response()
plt.figure()
plt.grid()
plt.plot(T, y_out.T)
plt.show()
```
