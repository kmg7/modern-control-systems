```python
import control as ct
import numpy as np
import display_utils as dp
import matplotlib.pyplot as plt
```

```python
sysg = ct.tf([1], [1, 1, 0])
dp.tf(sysg, "G(s)")

K1 = 100; K2 = 20
num1 = [11, K1]; num2 = [11, K2]
den = [0, 1]
sys1 = ct.tf(num1, den)
sys2 = ct.tf(num2, den)

sysa = ct.feedback(sysg, sys1)
dp.tf(sysa, "Closed-loop with K=100")
sysb = ct.feedback(sysg, sys2)
dp.tf(sysb, "Closed-loop with K=20")

sysc = ct.minreal(sysa)
dp.tf(sysc, "Reduced Closed-loop with K=100")
sysd = ct.minreal(sysb)
dp.tf(sysd, "Reduced Closed-loop with K=20")

t = np.arange(0, 2.5, 0.01)
_, y1 = sysc.step_response(t) 
_, y2 = sysd.step_response(t)
plt.plot(t, y1, label="K=100")
plt.plot(t, y2, label="K=20")
plt.title("Step Responses for Different K values")
plt.xlabel("Time (s)")
plt.ylabel("Response")
plt.legend()
plt.grid()
plt.show()
```
