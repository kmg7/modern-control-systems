```python
import control as ct
import numpy as np
import display_utils as dp
import matplotlib.pyplot as plt
```

```python
R = 1.0    # ohm
L = 0.5    # H
C = 1.0    # F

# H(s) = 1 / (L*C*s^2 + R*C*s + 1)
num = [1.0]
den = [L * C, R * C, 1.0]
tf = ct.tf(num, den)
dp.tf(tf)
ss = ct.ss(tf)
dp.ss(ss)
```

```python
T, y_out = ss.step_response()
plt.figure()
plt.plot(T, y_out.T)
plt.title("Step Response")
plt.xlabel("Time (s)")
plt.ylabel("Output Voltage (V)")
plt.grid()
plt.show()
```
