```python
import control as ct
from control.matlab import lsim
import numpy as np
import display_utils as dp
import matplotlib.pyplot as plt
```

```python
A = np.array([[0, 1], [-4, -7]])
B = np.array([[0], [1]])
C = np.array([[1, 0]])
D = np.array([[0]])

ss = ct.ss(A, B, C, D)
dp.ss(ss)

x0 = np.array([[1], [0]])
t = np.arange(0, 10, 0.01)
u = np.zeros((len(t), 1))

y_out, T, x_out = lsim(ss, U=u, T=t, X0=x0)
plt.figure()
plt.plot(T, y_out, label='y')
plt.plot(T, x_out[:, 0], '--', label='x1')
plt.plot(T, x_out[:, 1], label='x2')
plt.grid()
plt.legend()
plt.show()
```
