```python
import control as ct
import numpy as np
import matplotlib.pyplot as plt

def tf_of_system(z):
    num = [20/z, 20]
    denum = [1, 3, 20]
    return ct.tf(num, denum)

t = np.arange(0, 5, 0.01)

out_5 = tf_of_system(5.0).step_response(t).outputs
out_10 = tf_of_system(10.0).step_response(t).outputs
out_15 = tf_of_system(15.0).step_response(t).outputs

plt.figure(figsize=(10, 8))
plt.plot(t, out_5, label='z = 5.0', linestyle='-')
plt.plot(t, out_10, label='z = 10.0', linestyle='--')
plt.plot(t, out_15, label='z = 15.0', linestyle=':')
plt.grid(True)
plt.xlabel('Time (sec)')
plt.ylabel('System Output')
plt.legend()
plt.show()
```
