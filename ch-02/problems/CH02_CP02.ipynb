```python
import numpy as np
import control as ct
import matplotlib.pyplot as plt
controller = ct.tf([1], [1,1])
plant = ct.tf([1,2], [1,3])
feedback = 4
```

```python
# a
sys = ct.feedback(ct.series(controller, plant), feedback, sign= -1)
print(sys)
```

```python
# b
T, yout = sys.step_response(T=np.linspace(0, 3, 200))
print(yout[-1])
plt.plot(T, yout)
```
