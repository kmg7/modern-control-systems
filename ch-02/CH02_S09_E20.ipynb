```python
import control as ct
import matplotlib.pyplot as plt
import numpy as np

sys = ct.tf([5400], [2, 2.5, 5402])
print(sys)

t = np.arange(0, 3, 0.005)
T, yout = ct.step_response(sys, t)
plt.plot(T, yout)
plt.title('Step Response of the Traction Motor Wheel Velocity Control System')
plt.xlabel('Time [s]')
plt.ylabel('Wheel velocity [rad/s]')
```
